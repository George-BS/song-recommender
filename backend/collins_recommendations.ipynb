```python
import pandas as pd
import joblib
from pathlib import Path
```

```python
BASE_DIR = Path.cwd()

voyage_df = pd.DataFrame(
    joblib.load(BASE_DIR / "debug/6Db1zZcoO7JPczkjjU4vuX" / "songs.pkl")
)
```

```python
voyage_df.head()
```

```python
# Get IDs of all liked songs
liked_ids = list(voyage_df["track_id"].values)
```

```python
# Load all songs to check how many in the taste vector are available
# Why -> Tast vector is missing all song features - We need to impute with known values

import joblib

songs_df = joblib.load("songs.pkl")
```

```python
songs_df.head()
```

```python
liked_features = songs_df[songs_df["track_id"].isin(liked_ids)]
liked_features.head()
```

```python
# Extract numeric features
num_feat= liked_features.select_dtypes(include='number').columns.to_list()
```

```python
# Import the model
model = joblib.load("model.pkl")

# Get predictions from numeric features in the taste vector
distances, indices = model.kneighbors(liked_features[num_feat])
```

```python
indices
```

```python
recommendations = []
for i in indices:
    recommendations.extend(i)
```
